# How Progabide Faded: A Timeline of Efficacy, Safety, and Used


### **Final Project Notebook**

This notebook documents the final product created for the Progabide analysis project, including the visual timeline graphic, description, and a full resource manifest. The title of the infographic was inspired by a visual investigation of why Progabide, an early GABAergic anti-seizure drug, saw early scientific enthusiasm but eventually declined in clinical and regulatory standing. An image of the Final Infographic is included below, followed by a detailed description of the timeline and a comprehensive list of resources used in its creation.

## 3. Description of the Final Product

The final product is a **comprehensive visual timeline** illustrating the regulatory, clinical, and scientific trajectory of **Progabide**, an early antiepileptic drug. The epilepsy.csv which can be found in the setup folder is from  https://www.kaggle.com/code/ukveteran/compedjma-epilepsy/input. All other sources are listed below.

### **What the final product shows:**
- Regulatory decline (1985–2020)
- Key historical milestones
- Effectiveness and safety context
- Demographic data of the clinical study
- Comparative seizure freedom rates
